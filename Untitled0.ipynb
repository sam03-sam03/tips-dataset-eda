```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

```python
tips['tip_pct'] = (tips['tip'] / tips['total_bill']) * 100
tips.head()
```

```python
sns.barplot(data=tips, x='day', y='tip_pct')
```

```python
tips.groupby('day')['tip_pct'].mean()
```

```python
tips.groupby('size')['tip_pct'].mean()
```

```python
sns.barplot(data=tips, x='size', y='tip_pct')
```

```python
tips['size'].value_counts()
```

just did an analysis on 244 restaurant bills to understand tipping patterns. Found out that Friday has the highest average tip percentage (16.99 %) while Saturday has the lowest (15.31%) , despite Saturday typically being the busiest night.Also found that tip percentage decreases as group size increases,suggesting larger groups don't tip proportionally to their bill size.
